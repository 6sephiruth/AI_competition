```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
boston['feature_names']
```

```python
boston_house_df = pd.DataFrame(boston['data'])
boston_house_df
```

```python
boston_house_df = boston_house_df.rename(columns={0:'CRIM', 1:'ZN', 2:'INDUS'})
```

## 1에 가까운 값: 두 변수들간의 양의 상관관계가 있음.
## 0에 가까운 값: 두 변수들간의 상관 관계가 없음.
## -1에 가까운 값: 두 변수들간의 음의 상관 관계가 있음.

### 예를 들어서 범죄율이 높은데랑 가격은 -1에 가깝겠지?

```python
corr = boston_house_df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr,
            vmax=0.8,
            linewidths=0.01,
            square=True,
            annot=True,
            cmap='YlGnBu')
plt.title('Correlation Matrix')
```

```python
boston_data_df = pd.DataFrame(boston['data'])
boston_target_df = pd.DataFrame(boston['target'])
boston_target_df = boston_target_df.rename(columns={0:'PRICE'})
boston_full_df = pd.concat([boston_data_df, boston_target_df], axis=1)
boston_full_df = boston_full_df.rename(columns={5:'RM'})
boston_full_df
```

```python
sns.regplot(data=boston_full_df, x='RM', y='PRICE')
plt.show()
```
